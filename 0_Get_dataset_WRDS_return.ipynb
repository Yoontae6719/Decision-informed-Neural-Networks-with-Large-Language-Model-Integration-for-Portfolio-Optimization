We added the macro variable (foucsd on Composite Leading Indicators Index) such as 
- ICSA : https://fred.stlouisfed.org/series/ICSA
- UMCSENT : https://fred.stlouisfed.org/series/UMCSENT
- HSN1F : https://fred.stlouisfed.org/series/HSN1F
- UNRATE : https://fred.stlouisfed.org/series/UNRATE
- HIgh-Yield Bond Spread : https://fred.stlouisfed.org/series/BAMLH0A0HYM2

```python
import FinanceDataReader as fdr
```

```python
ICSA = pd.DataFrame({"ICSA" : fdr.DataReader('FRED:NASDAQCOM,ICSA', start='2010', end='2024-09-01')["ICSA"]}).reset_index()
UMCSENT = pd.DataFrame({"UMCSENT" : fdr.DataReader('FRED:NASDAQCOM,UMCSENT', start='2010', end='2024-09-01')["UMCSENT"]}).reset_index()
HSN1F = pd.DataFrame({"HSN1F" : fdr.DataReader('FRED:NASDAQCOM,HSN1F', start='2010', end='2024-09-01')["HSN1F"]}).reset_index()
UNRATE = pd.DataFrame({"UNRATE" : fdr.DataReader('FRED:NASDAQCOM,UNRATE', start='2010', end='2024-09-01')["UNRATE"]}).reset_index()
HYBS = pd.DataFrame({"HYBS" :  fdr.DataReader('FRED:NASDAQCOM,BAMLH0A0HYM2', start='2010', end='2024-09-01')["BAMLH0A0HYM2"]}).reset_index()
```

```python
ICSA.columns = ["Date", "ICSA"]
UMCSENT.columns = ["Date", "UMCSENT"]
HSN1F.columns = ["Date", "HSN1F"]
UNRATE.columns = ["Date", "UNRATE"]
HYBS.columns = ["Date", "HYBS"]
```

```python
ICSA = ICSA[ICSA["Date"].isin(list(SNP["Date"]))]
UMCSENT = UMCSENT[UMCSENT["Date"].isin(list(SNP["Date"]))]
HSN1F = HSN1F[HSN1F["Date"].isin(list(SNP["Date"]))]
UNRATE = UNRATE[UNRATE["Date"].isin(list(SNP["Date"]))]
HYBS = HYBS[HYBS["Date"].isin(list(SNP["Date"]))]
```

```python
ICSA['Date'] = pd.to_datetime(ICSA['Date']).dt.strftime('%Y-%m-%d')
UMCSENT['Date'] = pd.to_datetime(UMCSENT['Date']).dt.strftime('%Y-%m-%d')
HSN1F['Date'] = pd.to_datetime(HSN1F['Date']).dt.strftime('%Y-%m-%d')
UNRATE['Date'] = pd.to_datetime(UNRATE['Date']).dt.strftime('%Y-%m-%d')
HYBS['Date'] = pd.to_datetime(HYBS['Date']).dt.strftime('%Y-%m-%d')
```

```python
merged_df = SNP.merge(ICSA, on="Date", how="outer") \
               .merge(UMCSENT, on="Date", how="outer") \
               .merge(HSN1F, on="Date", how="outer") \
               .merge(UNRATE, on="Date", how="outer") \
               .merge(HYBS, on="Date", how="outer")
```

```python
merged_df.shape[0] == SNP.shape[0]
```

```python
merged_df.fillna(method='ffill', inplace=True)
```
